```python
import wbdata
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
countries = ["IN", "PAK", "LKA"]

indicators = {'NY.GNP.PCAP.CD':'GNI per Capita'} # this is supposed to be the dictionary used to indicators to be passed

df = wbdata.get_dataframe(indicators, country = countries, convert_date = False)

dfu = df.unstack(level=0)

dfu.plot()
plt.legend(loc='best')
plt.title("GNI per Capita($USD, Atlas Method)")
plt.xlabel('Date')
plt.ylabel("GNI per Capita($USD, Atlas Method)")
```
